```python
def quicksort(L, reverse=False):
    if len(L) < 2:
        return L

    else:
        pivot_idx = len(L) // 2
        pivot = L[pivot_idx]

        less, equal, greater = [], [], []

        for item in L:
            if item < pivot:
                less.append(item)
            elif item == pivot:
                equal.append(item)
            else:
                greater.append(item)

        if not reverse:
            return quicksort(less, reverse) + equal + quicksort(greater, reverse)
        else:
            return quicksort(greater, reverse) + equal + quicksort(less, reverse)

print(quicksort([8, 0, 2, 6, 1, 7, 3], reverse=True))
```

* For your program, you must define and use the following functions:

  * calcAverage(alist): Calculates the average from the list
* Expansion compared to Assignment II: use try-except to avoid invalid inputs (e.g., negative values, string), invalid operations, etc

```python
def getValues(n: int) -> list:
  '''
  Creates and returns a list of weights

  Inputs:
  - n: number of items in the list

  Outputs:
  - A list
  '''
  items = []

  for i in range(n):
    try:
      item = float(input('Enter a weight value: '))
    except Exception as e:
      print(e)
      item = float('nan')

    items.append(item)
    print(items)
    return items

def calcAverage(alist: list) -> float:
  '''
  Returns the average of list alist

  Input:
  - alist: list of weights

  Ouput:
  - The average value of the list
  '''
  try:
    return sum(alist[2:]) / len(alist[2:])
  except Exception as e:
    print(e)
    return float('nan')

def main():
  num = int(input('How many weights?'))
  if num <= 0:
    raise ValueError('num must be positive')

weights = getValues(num)
avg = calcAverage(weights)
print(f'Average: {avg}lbs.')


main()
```

Greeting class
* Write a Python class and populate it with two attributes and methods
  * Attributes: first/last names
  * Method display(): takes a Boolean input parameter; displays the attribute values either formally (last, first) or informally (first)
  * Special method __str__()

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def __str__(self):
    return self.first + ' ' + self.last

  def display(self, reverse=True):
    if reverse:
      print(f'{self.last}, {self.first}')
    else:
      print(self.first)

name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse=False)
```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newname):
    self.__name = newname

  def __len__(self):
    return len(self.__courses)

  def enroll(self, course):
    if course in self.__courses:
      print('You have already enrolled in', course)
    else:
      self.__courses.append(course)

s = Student('Bob', 12345)

```

Classes Point and Rectangle
* Create a class Point that has:
  * Two coordinate attributes (x,y)
  * A method distance() that calculates the distance between two instances of the class
* Create a class Rectangle that has:
  * Six attributes: four Point instances and two attributes for width/height. The latter two are calculated based on pairs of the four Point instances.
  * Three methods for calculating width, height, area of the rectangle.

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance (self, other):
    return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** .5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(4,0)
    self.p3 = Point(4,4)
    self.p4 = Point(0,4)
    self.width = self.p1.distance(self.p2)
    self.height = self.p2.distance(self.p3)

def area(self):
  return self.width * self.height
```
